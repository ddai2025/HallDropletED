```python
from DropletSystem import *

# INPUT
N = 6
L_tot = 75
N_workers = 48
interaction = -1
cycles = 1

# Initialize disk system
suffix = interaction
system = Droplet(N, L_tot, N_workers, interaction, suffix)
system.build_H()

# Build basis for L_CM = 0
thresh = 1e-24
l_seq = build_l_seq(system.L_ex, thresh)
system.build_L_CM_basis(l_seq, cycles)

# Diagonalize and save
system.diagonalize(10 ** 9)
system.save_data()
```
